```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings('ignore')
```

```python
GPS = pd.read_excel('DATA.xlsx', sheet_name='DATOS')
GPS
```

```python
unidades = GPS['Descripción'].unique()
unidades
```

```python
punidad = GPS[GPS['Descripción'] == unidades[0]]
#Listo
```

```python
punidad['Fecha y Hora'] = punidad['Fecha'].dt.strftime('%Y-%m-%d') + ' ' + punidad['Hora'].astype(str)
#Listo
```

```python
punidad = punidad.drop(columns = ['Fecha', 'Hora','Dirección', 'Altitud', 'Ángulo', 'Odómetro', 'Unidad', 'Diferencia', 'Estado'], axis = 1)
#Listo
```

```python
punidad['Fecha y Hora'] = pd.to_datetime(punidad['Fecha y Hora'], format='%Y-%m-%d %H:%M:%S')
#Listo
```

```python
punidad ['dtime'] = punidad['Fecha y Hora'].diff().dt.total_seconds()
#Listo
```

```python
mask = (punidad['Ignición'] == 'Encendido') & (punidad['Ignición'].shift(1) == 'Encendido')
punidad = punidad[mask].reset_index(drop=True)
#Listo
```

```python
dtotal = pd.DataFrame(columns=['Descripción','Latitud','Longitud','Velocidad','Ignición','Fecha y Hora','dtime'])
for i in unidades:
    punidad = GPS[GPS['Descripción'] == i]
    punidad['Fecha y Hora'] = punidad['Fecha'].dt.strftime('%Y-%m-%d') + ' ' + punidad['Hora'].astype(str)
    punidad = punidad.drop(columns = ['Fecha', 'Hora','Dirección', 'Altitud', 'Ángulo', 'Odómetro', 'Unidad', 'Diferencia', 'Estado'], axis = 1)
    punidad['Fecha y Hora'] = pd.to_datetime(punidad['Fecha y Hora'], format='%Y-%m-%d %H:%M:%S')
    punidad ['dtime'] = punidad['Fecha y Hora'].diff().dt.total_seconds()
    mask = (punidad['Ignición'] == 'Encendido') & (punidad['Ignición'].shift(1) == 'Encendido')
    punidad = punidad[mask].reset_index(drop=True)
    #dtotal = dtotal.append(punidad, ignore_index = True)  
    dtotal = pd.concat([dtotal, punidad], ignore_index=True)
```

```python
dtotal.loc[dtotal['dtime'] > 5, 'dtime'] = 6
```

```python
dtotal.groupby('dtime')['Descripción'].count()
```

```python
informe = pd.concat([dtotal['Descripción'], pd.get_dummies(dtotal.dtime, dtype=int)],axis=1)
```

```python
informe = informe.groupby('Descripción').sum()
informe = informe.reset_index()
informe
```
